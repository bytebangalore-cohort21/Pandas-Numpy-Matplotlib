```python
df_seattle = pd.read_excel("Seattle_dataset.xlsx")
df_seattle.sample()
```

```python
import datetime as dt

def convertdatetime(date):
    return dt.datetime.strptime(str(date), '%Y%m%d')
```

```python
df_seattle['new_date'] = df_seattle['DATE'].apply(convertdatetime)
```

```python
df_seattle.pivot_table(index='STATION_NAME',columns='new_date',values='TMIN').transpose().plot()
```
